## process the titles 

* convert OK, Okay, ok, okay to OK
* fill the missing lyrics of a music using the title of the music, search the title in the lyrics and fill the missing lyrics
* make them lowercase

```python
music['title'] = music['title'].str.lower()
music['artist'] = music['artist'].str.lower()
music['lyric'] = music['lyric'].str.lower()
```

```python
music
```

```python
def convert_OK(text):
    text = text.split()
    for i in range(len(text)):
        if text[i] == 'ok' or text[i] == 'okay' or text[i] == 'ok.' or text[i] == 'okay.' or text[i] == 'ok,' or text[i] == 'okay,' or text[i] == 'ok!' or text[i] == 'okay!' or text[i] == 'ok?' or text[i] == 'okay?':
            text[i] = 'ok'
    return ' '.join(text)
```

```python
music['title'] = music['title'].apply(str)
```

```python
music['title'] = music['title'].apply(convert_OK)
```

### fill the missing lyrics of a music using the title of the music

```python
nan_lyric = music[music['lyric'].isnull() & music['title'].notnull()]
nan_lyric
```

```python
full_lyric = music[music['lyric'].notnull()]
```

```python
import pandas as pd

# Separate rows with and without missing lyrics
nan_lyric = music[music['lyric'].isnull() & music['title'].notnull()]
full_lyric = music[music['lyric'].notnull()]

# Create a DataFrame with title and lyric (dropping duplicates for efficient merge)
full_lyric_mapping = full_lyric[['title', 'lyric']].drop_duplicates()

# Merge `nan_lyric` with `full_lyric_mapping` on title to get the missing lyrics
nan_lyric_filled = nan_lyric.merge(full_lyric_mapping, on='title', how='left', suffixes=('', '_fill'))

# Use fillna to update the original lyrics with the matched lyrics
music['lyric'] = music['lyric'].fillna(nan_lyric_filled['lyric_fill'])

# Drop the temporary column if it was added
music.drop(columns=['lyric_fill'], inplace=True, errors='ignore')
```

# drop the non-latin characters

```python
import re

def is_english_text(text):
    """
    Check if a text contains predominantly English characters.
    Returns True if the text is mostly English, False otherwise.
    """
    # Regex to allow only English letters, numbers, and common punctuation
    english_regex = re.compile(r'^[\x00-\x7F]+$')  # Matches ASCII characters only
    return bool(english_regex.match(text))

# Sample list of tweets including Indian scripts
tweets_ = [
    "This is an English tweet!",
    "यह हिंदी में एक ट्वीट है",  # Hindi
    "123456789",  # Numbers only
    "Another English text with numbers 123!",
    "ఇది తెలుగు ట్వీట్",  # Telugu
]

# Filter only English texts
filtered_tweets = [tweet for tweet in tweets_ if is_english_text(tweet)]

print("Filtered Tweets:", filtered_tweets)
```

```python
music['lyric'] = music['lyric'].apply(str)
music.loc[music['lyric'].apply(is_english_text)]
```

```python
music = music[music['lyric'].apply(is_english_text)]
```

```python
# Find the unique user_ids in both datasets
music_user_ids = set(music['user_id'])
tweets_user_ids = set(tweets['user_id'])

# Find the intersection of user_ids
matching_user_ids = music_user_ids.intersection(tweets_user_ids)

# Count the number of matches
num_matches = len(matching_user_ids)

print(f"Number of matching user_ids: {num_matches}")
```

```python
matching_musics = music[music['user_id'].isin(matching_user_ids)]
```

```python
matching_musics.disorder.value_counts()
```

```python
plt.figure(figsize=(10, 6))
matching_musics.disorder.value_counts().plot(kind='bar')
plt.title('Disorder Distribution in Matching Users')
plt.show()
```

```python
matching_musics.to_csv('../data/cleaned_matching_disorder_musics.csv', index=False)
```

```python
import pandas as pd

# Assume `df` is your DataFrame
df_sampled = []

# Group by disorder
grouped = matching_musics.groupby('disorder')

for disorder, group in grouped:
    # Ensure all unique user_ids are included
    unique_user_ids = group.groupby('user_id').head(1)
    
    # Calculate remaining samples needed to reach 500
    remaining = max(0, 500 - len(unique_user_ids))
    
    # Randomly sample the remaining rows (excluding already selected ones)
    remaining_samples = group.drop(unique_user_ids.index).sample(n=remaining, random_state=42, replace=False)
    
    # Combine unique user_id samples and additional random samples
    sampled_group = pd.concat([unique_user_ids, remaining_samples])
    
    # Append to the result
    df_sampled.append(sampled_group)

# Combine all sampled groups
df_sampled = pd.concat(df_sampled).reset_index(drop=True)

print(f"Total sampled songs: {len(df_sampled)}")
```
